```python
import warnings
warnings.filterwarnings("ignore")
from transformers import pipeline

# Load summarizer pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Prompt user for input
print("Enter a long educational article or paragraph:")
text = input()

# Truncate input if it's too long for the model
max_input_length = 1024  # BART's token limit
if len(text) > max_input_length:
    text = text[:max_input_length]

# Generate summary
summary = summarizer(text, max_length=150, min_length=25, do_sample=False)

# Print summary
print("\nüìù Summary:\n", summary[0]['summary_text'])
```

```python
import warnings
warnings.filterwarnings("ignore")
from transformers import pipeline

# Load summarization pipeline with a valid T5 model
summarizer = pipeline("summarization", model="t5-base")

# Take user input
text = input("Enter your input: ")

# T5 requires task prefix
formatted_text = "summarize: " + text

# Generate summary
summary = summarizer(formatted_text, max_length=150, min_length=25, do_sample=False)

# Output
print("\nüìù Summary:\n", summary[0]['summary_text'])
```
